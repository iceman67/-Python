```python
import pandas as pd

weatherFrame = pd.read_csv('weather_232_2021.csv', encoding='cp949', header=0,index_col = 0)
TgFrame = pd.read_csv('TG_108_2021.csv',encoding='cp949', header=0,index_col = 0)
```

```python
TgFrame
```

```python
weatherFrame.dtypes
```

```python
TgFrame.dtypes
```

```python
DataFrame =  pd.merge(weatherFrame, TgFrame,how = "outer",on='tm')
DataFrame
```

# 불필요 열 제거

```python
DataFrame.info
```

```python
if 'stnId' in DataFrame.columns :
    DataFrame = DataFrame.drop(['stnId'],axis=1)
DataFrame
```

```python
if 'gateway' in DataFrame.columns :
    DataFrame = DataFrame.drop(['gateway'],axis=1)
DataFrame
```

```python
if 'code' in DataFrame.columns :
    DataFrame = DataFrame.drop(['code'],axis=1)
DataFrame
```

# GroupBy

```python
DataFrame.info
```

```python
DataFrame.dtypes
```

```python
DataFrame['dateTime'] = DataFrame.index
DataFrame
DataFrame.dtypes
```

```python
DataFrame['dateTime'] = DataFrame['dateTime'].astype(str)
```

```python
DataFrame['dateTime'] = pd.to_datetime(DataFrame['dateTime'])
```

```python
DataFrame.dtypes
```

```python
DataFrame
```

```python
DataFrame.index = DataFrame['dateTime']
```

```python
DataFrame
```

```python
if 'dateTime' in DataFrame.columns :
    DataFrame = DataFrame.drop(['dateTime'],axis=1)
DataFrame
```

```python
forGroupping = DataFrame
forGroupping['month'] = pd.DatetimeIndex(DataFrame.index).month
newDataFrame = forGroupping.groupby('month').mean()
newDataFrame
```

# 시각화

```python
import matplotlib.pyplot as plt
```

데이터 수정없이 평균/최저/최고 기온과 교통이동량의 월별 평균치를 시각화했을 때

```python
newDataFrame.plot()
plt.show()
```

평균기온 월별 평균 시각화 그래프

```python
plt.plot(newDataFrame['avgTa'],
         color='skyblue',
         marker='o', markerfacecolor='blue')
plt.xlabel('Month')
plt.ylabel('Temperature(avgTa)')
```

최저기온 월별 평균 시각화 그래프

```python
plt.plot(newDataFrame['minTa'],
         color='yellow',
         marker='o', markerfacecolor='yellow')
plt.xlabel('Month')
plt.ylabel('Temperature(minTa)')
```

최고기온 월별 평균 시각화 그래프

```python
plt.plot(newDataFrame[' maxTa'],
         color='green',
         marker='o', markerfacecolor='green')
plt.xlabel('Month')
plt.ylabel('Temperature(maxTa)')
```

평균/최저/최고기온 월별 평균 시각화 그래프

```python
plt.plot(newDataFrame['avgTa'],
         color='skyblue',
         marker='o', markerfacecolor='blue')
plt.plot(newDataFrame['minTa'],
         color='yellow',
         marker='o', markerfacecolor='yellow')
plt.plot(newDataFrame[' maxTa'],
         color='green',
         marker='o', markerfacecolor='green')
plt.xlabel('Month')
plt.ylabel('Temperature')
```

교통이동량 월별 평균 시각화 그래프

```python
plt.plot(newDataFrame['totalTRF'],
         color='red',
         marker='o', markerfacecolor='red', label ="TRF")
plt.plot(newDataFrame['avgTa'],
         color='skyblue',
         marker='o', markerfacecolor='blue', label ="avgTa")
plt.yscale("log")
plt.xlabel('Month')
plt.ylabel('Traffic(totalTRF)')
plt.legend()
```

```python
plt.plot(newDataFrame['totalTRF'],
         color='red',
         marker='o', markerfacecolor='red')

plt.plot
plt.xlabel('Month')
plt.ylabel('Traffic(totalTRF)')
```
