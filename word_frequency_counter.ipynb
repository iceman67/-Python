```python
url ="http://www.daum.net"
page = requests.get(url)
doc = page.text.lower()
```

```python
words = re.findall(r"\w+", doc)
print (Counter(words).most_common(15))
freq= Counter(words).most_common(15)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame(freq,  columns=['Word', 'Freq'])
df = df.set_index("Word")
df.plot(kind="bar")
plt.show()
```
